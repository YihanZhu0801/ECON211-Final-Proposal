```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```
share_price=pd.read_csv('https://raw.githubusercontent.com/YihanZhu0801/ECON211-Final-Proposal/main/SharePrice.csv')
share_price.head()
```

```
sp_x = share_price['Dates']
sp_y = share_price['3M Company (NYSE:MMM) - Share Pricing']


#plt.figure()    
fig = plt.figure(figsize=(20,5))
plt.plot(sp_x, sp_y) 
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(50))
plt.xticks(rotation=90)
plt.xlabel('Dates')
plt.ylabel('Share Price(USD)')
plt.title('Daily Share Price of 3M Company from 2005 to 2015')
plt.show()
```

```
share_price_Dec2016=pd.read_csv('https://raw.githubusercontent.com/YihanZhu0801/ECON211-Final-Proposal/main/SharePriceDec2016.csv')
share_price_Dec2016.head()
```
